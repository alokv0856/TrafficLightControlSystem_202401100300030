```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import time
```

```python
# Upload dataset
uploaded = files.upload()

# Load dataset
df = pd.read_csv("/content/traffic_data.csv")

# Display first few rows
print("Dataset Preview:")
print(df.head())

# Basic info about dataset
print("\nDataset Info:")
print(df.info())

# Summary statistics
print("\nStatistical Summary:")
print(df.describe())
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='Date', y='BounceRate', data=df, marker='o', color='red')
plt.xticks(rotation=45)
plt.title("Bounce Rate Over Time")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='PageViews', y='BounceRate', data=df, color='purple')
plt.title("Page Views vs Bounce Rate")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='UniqueVisitors', y='BounceRate', data=df, color='blue')
plt.title("Unique Visitors vs Bounce Rate")
plt.show()
```
